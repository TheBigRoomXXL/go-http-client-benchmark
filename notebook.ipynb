```python
import numpy as np
import csv
import matplotlib.pyplot as plt 
from pathlib import Path
```

```python
csvpath = Path("results.localhost.csv")
with csvpath.open() as csvfile:
    reader = csv.reader(csvfile)
    data = list(reader)
    data = np.array(data[1:],dtype=float)
```

```python
fig, ax = plt.subplots()
ax.plot(data[:,3],data[:,4], ".", alpha=0.5)
ax.hlines(3000,0,2,colors="y",linestyles="dashed")
ax.hlines(2000,0,2,colors="r",linestyles="dashed")
```

```python
fig, ax = plt.subplots()
ax.plot(data[:,1],data[:,4], ".", alpha=0.5)
ax.hlines(3000,0,1000,colors="y",linestyles="dashed")
ax.hlines(2000,0,1000,colors="r",linestyles="dashed")
```

```python
fig, ax = plt.subplots()
ax.plot(data[:,2],data[:,4], ".", alpha=0.5)
ax.hlines(3000,0,1500,colors="y",linestyles="dashed")
ax.hlines(2000,0,1500,colors="r",linestyles="dashed")
```

```python
myhist_sum, _, _ = np.histogram2d(data[:,3],data[:,2],bins=15, weights=data[:,4])
myhist_count, _, _ = np.histogram2d(data[:,3],data[:,2],bins=15)
myhist = myhist_sum/myhist_count

fig, ax = plt.subplots()
ax.imshow(myhist)
```
