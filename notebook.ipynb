```python
import numpy as np
import csv
import matplotlib.pyplot as plt 
from pathlib import Path
```

```python
csvpath = Path("results/batch-1/public-network.csv")
with csvpath.open() as csvfile:
    reader = csv.reader(csvfile)
    data = list(reader)
    data = np.array(data[1:],dtype=float)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("Latency (s)")
ax.set_ylabel("Throughput (req/s)")
ax.plot(data[:,3],data[:,4], ".", alpha=0.5)
ax.hlines(3000,0,1,colors="y",linestyles="dashed")
ax.hlines(2000,0,1,colors="r",linestyles="dashed")
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("Response Size (KB)")
ax.set_ylabel("Throughput (req/s)")
ax.plot(data[:,1],data[:,4], ".", alpha=0.5)
ax.hlines(3000,0,500,colors="y",linestyles="dashed")
ax.hlines(2000,0,500,colors="r",linestyles="dashed")
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("Number of Threads")
ax.set_ylabel("Throughput (req/s)")
ax.plot(data[:,2],data[:,4], ".", alpha=0.5)
ax.hlines(3000,200,1000,colors="y",linestyles="dashed")
ax.hlines(2000,200,1000,colors="r",linestyles="dashed")
```

```python
localhost_csv_path = Path("results/batch-2/localhost.csv")
with localhost_csv_path.open() as csvfile:
    reader = csv.reader(csvfile)
    localhost_data = list(reader)
    localhost_data = np.array(localhost_data[1:],dtype=float)

public_csv_path = Path("results/batch-2/public-network.csv")
with public_csv_path.open() as csvfile:
    reader = csv.reader(csvfile)
    public_data = list(reader)
    public_data = np.array(public_data[1:],dtype=float)


fig, ax = plt.subplots(3,2,figsize=(12,15))
fig.suptitle("Throughput Variation", fontsize=26)

ax[0][0].set_title("Localhost", fontsize=20)
ax[0][0].set_xlabel("Latency (s)")
ax[0][0].set_ylabel("Throughput (req/s)")
ax[0][0].plot(localhost_data[:,3],localhost_data[:,4], ".", alpha=0.5)
ax[0][0].hlines(2000,0,8.5,colors="y",linestyles="dashed")

ax[1][0].set_xlabel("Response Size (KB)")
ax[1][0].set_ylabel("Throughput (req/s)")
ax[1][0].plot(localhost_data[:,1],localhost_data[:,4], ".", alpha=0.5)
ax[1][0].hlines(2000,0,500,colors="y",linestyles="dashed")

ax[2][0].set_xlabel("Number of Threads")
ax[2][0].set_ylabel("Throughput (req/s)")
ax[2][0].plot(localhost_data[:,2],localhost_data[:,4], ".", alpha=0.5)
ax[2][0].hlines(2000,200,1000,colors="y",linestyles="dashed")

ax[0][1].set_title("Public Network", fontsize=20)
ax[0][1].set_xlabel("Latency (s)")
ax[0][1].set_ylabel("Throughput (req/s)")
ax[0][1].plot(public_data[:,3],public_data[:,4], ".", alpha=0.5)
ax[0][1].hlines(2000,0,8.5,colors="y",linestyles="dashed")
ax[0][1].sharex(ax[0][0])
ax[0][1].sharey(ax[0][0])

ax[1][1].set_xlabel("Response Size (KB)")
ax[1][1].set_ylabel("Throughput (req/s)")
ax[1][1].plot(public_data[:,1],public_data[:,4], ".", alpha=0.5)
ax[1][1].hlines(2000,0,500,colors="y",linestyles="dashed")
ax[1][1].sharex(ax[1][0])
ax[1][1].sharey(ax[1][0])

ax[2][1].set_xlabel("Number of Threads")
ax[2][1].set_ylabel("Throughput (req/s)")
ax[2][1].plot(public_data[:,2],public_data[:,4], ".", alpha=0.5)
ax[2][1].hlines(2000,200,1000,colors="y",linestyles="dashed")
ax[2][1].sharex(ax[2][0])
ax[2][1].sharey(ax[2][0])
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))


data = public_data[public_data[:, 0] == 100]
dataA = data[data[:, 2] == 200]
dataB = data[data[:, 2] == 400]
dataC = data[data[:, 2] == 600]
dataD = data[data[:, 2] == 800]
dataE = data[data[:, 2] == 1000]

fig.suptitle("Response size impact on public network (delay=100ms)", fontsize=16)

ax[0].set_ylabel("Throughput (req/s)")
ax[0].set_xlabel("Size (KB)")

ax[0].plot(dataA[:, 1], dataA[:, 4], "-", label="200")
ax[0].plot(dataB[:, 1], dataB[:, 4], "-", label="400")
ax[0].plot(dataC[:, 1], dataC[:, 4], "-", label="600")
ax[0].plot(dataD[:, 1], dataD[:, 4], "-", label="800")
ax[0].plot(dataE[:, 1], dataE[:, 4], "-", label="1000")
ax[0].legend(title="Threads")


data = public_data[public_data[:,0] == 100]
dataA = data[data[:,2] == 200]
dataB = data[data[:,2] == 400]
dataC = data[data[:,2] == 600]
dataD = data[data[:,2] == 800]
dataE = data[data[:,2] == 1000]

# fig.suptitle("Latency vs Size (response delay = 100ms)")

ax[1].set_ylabel("Latency (s)")
ax[1].set_xlabel("Size (KB)")
ax[1].plot(dataA[:,1],dataA[:,3], "-", label="200")
ax[1].plot(dataB[:,1],dataB[:,3], "-", label="400")
ax[1].plot(dataC[:,1],dataC[:,3], "-", label="600")
ax[1].plot(dataD[:,1],dataD[:,3], "-", label="800")
ax[1].plot(dataE[:,1],dataE[:,3], "-", label="1000")
ax[1].legend(title="Threads")
```
